```python
import pandas as pd
import numpy as np
import folium
```

```python
GTFS_DIR = "gtfs-out"

trips = pd.read_csv(f"{GTFS_DIR}/trips.txt")
stop_times = pd.read_csv(f"{GTFS_DIR}/stop_times.txt")
shapes = pd.read_csv(f"{GTFS_DIR}/shapes.txt")
stops = pd.read_csv(f"{GTFS_DIR}/stops.txt")
```

# Get stop location

```python
stop_id = '1_11009'
stop_row = stops[stops.stop_id==stop_id]

stop_location = stop_row[['stop_lat', 'stop_lon']]
```

# Get one of the trips

```python
trip_id = stop_times[stop_times.stop_id==stop_id]['trip_id'].unique()[0]
```

# Get the shape_dist_traveled for our stop

```python
shape_dist_traveled = stop_times[(stop_times['stop_id']==stop_id) & (stop_times['trip_id']==trip_id)]['shape_dist_traveled'].iloc[0]
```

# Get the shape dists and points

```python
shape_id = trips[trips.trip_id==trip_id]['shape_id'].unique()[0]
shape_dists = shapes[shapes.shape_id==shape_id]['shape_dist_traveled'].to_numpy()
shape_pts = shapes[shapes.shape_id==shape_id][['shape_pt_lat', 'shape_pt_lon']].to_numpy()
```

# Get the shape point with the closest shape_dist_traveled

```python
closest_distance = shape_dists[np.abs(shape_dists - shape_dist_traveled).argmin()]
```

# Get the shape point

```python
street_point = shapes[shapes.shape_dist_traveled==closest_distance][['shape_pt_lat', 'shape_pt_lon']]
```

# Plot everything

```python
m = folium.Map(location=stop_location, zoom_start=50)

# shape points
for pt in shape_pts:
    folium.CircleMarker(
        pt,
        radius=3,
        color='green',
    ).add_to(m)

# gtfs stop location
folium.CircleMarker(
    stop_location,
    radius=3,
    color='blue',
).add_to(m)

# closest shape point
folium.CircleMarker(
    street_point,
    radius=3,
    color='red',
).add_to(m)

m
```
